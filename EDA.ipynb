data cleaning

```python
telco_data = telco_base_data.copy()
```

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors='coerce')
telco_data.isnull().sum()
```

```python
telco_data.loc[telco_data ['TotalCharges'].isnull() == True]
```

```python
#Removing missing values 
telco_data.dropna(how = 'any', inplace = True)
```

```python
# Get the max tenure
print(telco_data['tenure'].max()) 
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telco_data['tenure_group'] = pd.cut(telco_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
telco_data['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
telco_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
telco_data.head()
```


data exploration

univariant analysis(Plot distibution of individual predictors by churn)

```python
for i, predictor in enumerate(telco_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telco_data, x=predictor, hue='Churn')
    
    plt.title(f'Count of {predictor} by Churn')
    plt.xlabel(predictor)
    plt.ylabel('Count')
    plt.legend(title='Churn', loc='upper right')  # Optional: Customize legend
    plt.show()
```

```python
#Convert the target variable 'Churn' in a binary numeric variable
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes',1,0)
```

```python
telco_data.head()
```

```python
#Convert all the categorical variables into dummy variables
telco_data_dummies = pd.get_dummies(telco_data)
telco_data_dummies.head()
```

```python
#Relationship between Monthly Charges and Total Charges
plt.figure(figsize=(10, 6))  # Set figure size for better visibility
sns.scatterplot(data=telco_data_dummies, x='MonthlyCharges', y='TotalCharges')

plt.title('Relationship between Monthly Charges and Total Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.grid(True)  # Optional: Add grid for better readability
plt.show()
```

```python
#Churn by Monthly Charges and Total Charges

plt.figure(figsize=(10, 6))  # Set figure size for better visibility

# KDE plot for Monthly Charges by Churn status
sns.kdeplot(telco_data_dummies.MonthlyCharges[telco_data_dummies["Churn"] == 0],
            color="red", fill=True, label="No Churn")
sns.kdeplot(telco_data_dummies.MonthlyCharges[telco_data_dummies["Churn"] == 1],
            color="blue", fill=True, label="Churn")

# Adding labels and title
plt.ylabel('Density')
plt.xlabel('Monthly Charges')
plt.title('Monthly Charges by Churn')
plt.legend(loc='upper right')  

plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set figure size for better visibility

# KDE plot for Total Charges by Churn status
sns.kdeplot(telco_data_dummies.TotalCharges[telco_data_dummies["Churn"] == 0],
            color="red", fill=True, label="No Churn")
sns.kdeplot(telco_data_dummies.TotalCharges[telco_data_dummies["Churn"] == 1],
            color="blue", fill=True, label="Churn")

# Adding labels and title
plt.ylabel('Density')
plt.xlabel('Total Charges')
plt.title('Total Charges by Churn')
plt.legend(loc='upper right') 

plt.show()
```

```python
plt.figure(figsize=(20, 8))  # Set figure size for better visibility

# Calculate the correlation with 'Churn' and sort the values
correlation = telco_data_dummies.corr()['Churn'].sort_values(ascending=False)

# Create a bar plot
correlation.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.title('Correlation of Features with Churn', fontsize=20)
plt.xlabel('Features', fontsize=15)
plt.show()
```

```python
# Correlation Matrix

corr_matrix = telco_data_dummies.corr()
plt.figure(figsize = (15, 15))
sns.heatmap(corr_matrix, annot = True, fmt = '0.2f')
plt.title("Correlation Matrix of Telecom Customers", fontsize = 20)
plt.show()

# It is clearly shown that "voice_mail_plan" and "number_vmail_messages" are highly correlated.
# It is clearly shown that "total day charge" and "total daily minutes" are highly correlated.
```

bivariate analysis


```python
new_df1_target0=telco_data.loc[telco_data["Churn"]==0]
new_df1_target1=telco_data.loc[telco_data["Churn"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```
